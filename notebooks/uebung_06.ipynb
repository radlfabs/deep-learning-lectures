and it's derivative:

Ein sinnvolles Fehlerma ist die Kreuzentropie: $$H(\mathbf{X}, P, Q) = - \sum_{i=1}^{n}Q(x_i)\log_2 P(x_i)$$
Die Kreuzentropie ist ein Ma fr die hnlichkeit zwischen zwei Wahrscheinlichkeitsverteilungen. In unserem Fall sind $P$ und $Q$ die Wahrscheinlichkeitsverteilungen der tatschlichen und der vorhergesagten Klasse. Die Kreuzentropie ist minimal, wenn die beiden Verteilungen gleich sind.
